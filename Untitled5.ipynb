```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Function to convert duration to minutes
def convert_to_minutes(duration):
    if isinstance(duration, str):
        if not duration:
            return np.nan
        if 'h' in duration:
            try:
                hours, minutes = duration.split('h')
                minutes = minutes.replace('m', '').strip()
                total_minutes = int(hours) * 60 + int(minutes)
                return max(total_minutes, 0)
            except ValueError:
                return np.nan
        elif 'min' in duration:
            try:
                total_minutes = int(duration.replace(' min', '').strip())
                return max(total_minutes, 0)
            except ValueError:
                return np.nan
    return np.nan

df = pd.read_csv(path)
# Apply the function to the 'duration' column
df['duration'] = df['duration'].apply(convert_to_minutes)

# Now you can proceed with the analysis, grouping by 'mood' instead of 'genre'
avg_duration_mood = df.groupby('mood')['duration'].mean().sort_values(ascending=False)

# Create the barplot with switched axes
plt.figure(figsize=(12,15))
sns.barplot(y=avg_duration_mood.index, x=avg_duration_mood.values)

# Add title and y-tick rotation
plt.title('Average Movie Duration per Mood')
plt.xlabel('Duartion of the movies')
plt.ylabel('Mood')
plt.yticks(rotation=0)

# Display the plot
plt.show()

# Save the figure as a PNG file
plt.savefig('average_movie_duration_per_mood.png', bbox_inches='tight')  # Saves the plot in the current directory
```

```python
print("This bar graph is of the mood of the movie with the duration of the movie. It is observable that emotional, romantic movies have the longest of the durations, followed by mind - bending, bittersweet and adventure type movies. It is also noticed that Scary and horror related movies generally have the lowest of runtimes, it can be predicted from the graph that these movies almost of the time have around 100 or less than 100 mins of runtime, whereas if we take action movies into consideration it is noticed that they have varying runtimes as moods related to action such as adrenaline rush, violent, exciting movies are distributed all over many minutes. Some categories are repeated in the Y-axis to accomodate all the movies of that particular mood.")    
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv(path) 
df['release_year'] = pd.to_numeric(df['release_year'], errors = 'coerce') 
current_year = pd.to_datetime("now").year
df_last_10_years = df[df['release_year'] >= current_year - 10]
df_last_10_years.loc[:, 'cast'] = df_last_10_years['cast'].fillna('')
df_last_10_years.loc[:, 'cast'] = df_last_10_years['cast'].apply(lambda x: x.split(','))
actors_list = df_last_10_years['cast'].explode().value_counts()
top_10_actors = actors_list.head(10)
plt.figure(figsize=(20,10))
sns.barplot(x = top_10_actors.index, y = top_10_actors.values)
plt.title('Top 10 Actors with the Most Movies in the Last 10 Years')
plt.xlabel('Actor')
plt.ylabel('Number of movies')
plt.show()
```
